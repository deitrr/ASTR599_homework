Plot Discussion: The plot shows that the asteroid distribution has gaps at particular values of semi-major axis.  Theseare likely caused by mean motion resonances.  Families of orbits in particular ranges of semi-major axisnear these resonances will be unstable and thus no asteroids will be observed at these locations.
