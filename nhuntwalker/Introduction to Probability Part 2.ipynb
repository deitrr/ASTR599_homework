**Bivariate and multivariate distribution**

The distribution $h(x,y)$ should be interpreted as giving the probability that
$x$ is between $x$ and $x+dx$  _and_ that $y$ is between $y$ and $y+dy$.

The two variances are defined as

$V_x = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}  (x-\mu_x)^2 \, h(x,y)\,dx\,dy$

and

$V_y = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}  (y-\mu_y)^2 \, h(x,y)\,dx\,dy,$

where the mean values are defined as

$\mu_x= \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}  x \, h(x,y)\,dx\,dy$

and analogously for $\mu_y$.  In addition, the covariance of $x$ and $y$, which is a measure
of the dependence of the two variables on each other, is defined as

$V_{xy} = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}  (x-\mu_x)(y-\mu_y) \, h(x,y)\,dx\,dy.$

A very useful related result is that the variance of the sum $z=x+y$ is

$V_z = V_x + V_y + 2\,V_{xy}.$

When $x$ and $y$ are uncorrelated ($V_{xy}=0$), the variance of their sum is equal to the sum
of their variances. 

If $\sigma_{xy}=0$, then $x$ and $y$ are uncorrelated and we can treat them separately as two
independent one-dimensional distributions. 

$h(x,y) = h_x(x) \, h_y(y)$

**Bivariate Gaussians**

A generalization of the Gaussian distribution to the two-dimensional case is given by

$p(x,y|\mu_x, \mu_y, \sigma_x, \sigma_{y}, \sigma_{xy}) = 
       {1 \over 2 \pi \sigma_x \sigma_y \sqrt{1-\rho^2}} \exp{\left({-z^2 \over 2 (1-\rho^2)}\right)},$

where

$z^2 = {(x-\mu_x)^2 \over \sigma_x^2} +  {(y-\mu_y)^2 \over \sigma_y^2} 
       - 2\,\rho \,  {(x-\mu_x) \, (y-\mu_y) \over \sigma_x \, \sigma_y},$
       
and the (dimensionless) correlation coefficient between $x$ and $y$ is defined as

$\rho = {\sigma_{xy} \over \sigma_x \sigma_y}$


_The correlation in the data is defined by $\rho$_. For perfectly correlated variables such that $y=ax + b$, $\rho=a/|a|\equiv{\rm sign}(a)$, and for uncorrelated variables, $\rho=0$. 


The contours in the $(x,y)$ plane defined by $p(x,y|\mu_x, \mu_y, \sigma_x, \sigma_{y}, \sigma_{xy})$=constant
are ellipses centered on ($x=\mu_x$, $y=\mu_y$), and the angle
$\alpha$ (defined for $-\pi/2 \le \alpha \le \pi/2$)
between the $x$-axis and
the ellipses' major axis is given by

$\tan(2\alpha) =  2  \, \rho \, {\sigma_x \sigma_y  \over \sigma_x^2 - \sigma_y^2} = 
                               2  {\sigma_{xy} \over \sigma_x^2 - \sigma_y^2}.$
                               
_In matrix notation_

$p(x) = \frac{1}{\sqrt{2\pi} ||\Sigma||^{1/2} } \exp(- \frac{d^T \Sigma^{-1} d}{2})$

$\Sigma = \begin{bmatrix}\sigma_x^2 & \sigma_x\sigma_y \\ \sigma_x\sigma_y & \sigma_y^2\end{bmatrix}$

**Principal Axes**

When the $(x,y)$ coordinate system is rotated by an angle $\alpha$ around the point ($x=\mu_x$, $y=\mu_y$),

$P_1 = (x-\mu_x) \cos \alpha  + (y-\mu_y) \sin \alpha, \nonumber $

$P_2 = - (x-\mu_x) \sin \alpha + (y-\mu_y) \cos \alpha,$


the correlation between the two new variables $P_1$ and $P_2$ disappears, and the two widths are

$\sigma_{1,2}^2 = {\sigma_x^2 + \sigma_y^2 \over 2} \pm \sqrt{\left({\sigma_x^2 - \sigma_y^2 \over 2}\right)^2 + \sigma_{xy}^2}.$

The coordinate axes $P_1$ and $P_2$ are called the principal axes,
and $\sigma_1$ and $\sigma_2$ represent the minimum
and maximum widths obtainable for any rotation of the coordinate axes. I


Alternatively, starting from the principal axes frame, we can compute

$\sigma_x = \sqrt{\sigma_1^2 \cos^2 \alpha  + \sigma_2^2 \sin^2 \alpha  },$


$\sigma_y = \sqrt{\sigma_1^2 \sin^2 \alpha  + \sigma_2^2 \cos^2 \alpha  },$

and  ($\sigma_1 \ge \sigma_2$ by definition)

$\sigma_{xy} = (\sigma_1^2 - \sigma_2^2) \sin \alpha \cos \alpha.$

Note that $\sigma_{xy}$, and thus the correlation coefficient $\rho$, vanish for
both $\alpha=0$ and $\alpha=\pi/2$, and have maximum values for $\pi/4$.
By inverting, we get

$x = \mu_x + P_1 \, \cos \alpha  - P_2 \, \sin \alpha, \nonumber $

$y = \mu_y  + P_1 \, \sin \alpha + P_2 \, \cos \alpha.$

These expressions are very useful when generating mock samples based on bivariate Gaussians

**Estimating of a bivariate Gaussian distribution from data**

A bivariate Gaussian distribution is often encountered in practice when dealing
with two-dimensional problems. We can estimate  five parameters 
$(\overline{x}, \overline{y}, s_x, s_{y}, s_{xy})$ from the data. In particular, the correlation coefficient is estimated using Pearson's sample correlation coefficient, $r$. The principal axes can be easily found with $\alpha$ estimated using

$\tan(2\alpha) =  2  {s_x s_y  \over s_x^2 - s_y^2} \, r,$

where for simplicity we use the same symbol for both population and sample values of $\alpha$.

When working with real data sets  outliers can bias $\alpha$ because of the impact of outliers on $s_x$, $s_y$, and $r$. A simple and efficient remedy is to use the median instead of
the mean, and to use the interquartile
range to estimate variances.

To robustly estimate $r$, we can use 

$\rho = {V_u - V_w \over V_u + V_w}$

where $V$ is variance, and transformed coordinates are defined as (Cov($u$,$w$)=0 ie we are in principal axes space)
           
$u = {\sqrt{2}\over 2} \left( {x \over \sigma_x} + {y \over \sigma_y} \right)$

and

$w = {\sqrt{2}\over 2} \left( {x \over \sigma_x} - {y \over \sigma_y} \right).$

By substituting the robust estimator $\sigma_G^2$ in place of the variance $V$ 
we can compute a robust estimate of $r$, and in turn a robust estimate of the principal axis
angle $\alpha$.  

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
from matplotlib.patches import Ellipse
from astroML.stats import fit_bivariate_normal
from astroML.stats.random import bivariate_normal

# percent sign needs to be escaped if usetex is activated
import matplotlib
if matplotlib.rcParams.get('text.usetex'):
    pct = '\%'
else:
    pct = '%'

N = 1000

sigma1 = 2.0
sigma2 = 1.0
mu = [10, 10]
alpha_deg = 45.0
alpha = alpha_deg * np.pi / 180

#------------------------------------------------------------
# Draw N points from a multivariate normal distribution
#
#   we use the bivariate_normal function from astroML.  A more
#   general function for this is numpy.random.multivariate_normal(),
#   which requires the user to specify the full covariance matrix.
#   bivariate_normal() generates this covariance matrix for the
#   given inputs

np.random.seed(0)
X = bivariate_normal(mu, sigma1, sigma2, alpha, N)

#------------------------------------------------------------
# Create the figure showing the fits
fig = plt.figure(figsize=(10, 5))
fig.subplots_adjust(left=0.07, right=0.95, wspace=0.05,
                    bottom=0.1, top=0.95)


# We'll create two figures, with two levels of contamination
for i, f in enumerate([0.05, 0.15]):
    ax = fig.add_subplot(1, 2, i + 1)

    # add outliers distributed using a bivariate normal.
    X[:int(f * N)] = bivariate_normal((10, 10), 2, 5,
                                      45 * np.pi / 180., int(f * N))
    x, y = X.T

    # compute the non-robust statistics
    (mu_nr, sigma1_nr,
     sigma2_nr, alpha_nr) = fit_bivariate_normal(x, y, robust=False)

    # compute the robust statistics
    (mu_r, sigma1_r,
     sigma2_r, alpha_r) = fit_bivariate_normal(x, y, robust=True)

    # scatter the points
    ax.scatter(x, y, s=4, lw=0, c='k', alpha=0.5)

    # Draw elipses showing the fits
    for Nsig in [1, 3]:
        # True fit
        E = Ellipse((10, 10), sigma1 * Nsig, sigma2 * Nsig, alpha_deg,
                    lw=2, ec='k', fc='none')
        ax.add_patch(E)

        # Non-robust fit
        E = Ellipse(mu_nr, sigma1_nr * Nsig, sigma2_nr * Nsig,
                    (alpha_nr * 180. / np.pi),
                    lw=1, ec='k', fc='none', linestyle='dotted')
        ax.add_patch(E)

        # Robust fit
        E = Ellipse(mu_r, sigma1_r * Nsig, sigma2_r * Nsig,
                    (alpha_r * 180. / np.pi),
                    lw=2, ec='k', fc='none', linestyle='dashed')
        ax.add_patch(E)

    ax.text(0.02, 0.98, '%i%s outliers' % (f * 100, pct),
            ha='left', va='top', transform=ax.transAxes)

    ax.set_xlim(5.5, 14.5)
    ax.set_ylim(5.5, 14.5)
    ax.set_xlabel('x')

    # This is a bit of a hack:
    # We'll draw some lines off the picture to make our legend look better
    ax.plot([0], [0], '-k', lw=2, label='Input')
    ax.plot([0], [0], ':k', lw=2, label='Fit')
    ax.plot([0], [0], '--k', lw=2, label='Robust Fit')
    ax.legend(loc='lower right', prop=dict(size=12))

    if i == 0:
        ax.set_ylabel('y')
    else:
        ax.yaxis.set_major_formatter(plt.NullFormatter())

plt.show()
```

**Correlations and Correlation Coefficients**

Given two data sets of equal size $N$, $x$ and $y$, Pearson's sample correlation coefficient is

$r={\sum_{i=1}^N (x_i-\overline{x})(y_i-\overline{y})\over \sqrt{\sum_{i=1}^N (x_i-\overline{x})^2} \sqrt{\sum_{i=1}^N (y_i-\overline{y})^2}}$

with $-1 \le r \le 1$. For uncorrelated variables,
$r=0$. If the pairs ($x_i$, $y_i$) are drawn
from two uncorrelated univariate Gaussian distributions (ie the population correlation
coefficient $\rho=0$), then the distribution of $r$ follows Student's $t$ distribution with
$k=N-2$ degrees of freedom and

$t = r \sqrt{N-2  \over 1- r^2} $

Given this known distribution, a measured value of $r$ can be transformed into
the significance of the statement that $x$ and $y$ are correlated. For example,
if $N$=10, the probability that a value of $r$ as large as 0.72 would arise by chance
is 1% (the one-sided 99% confidence level for Student's $t$ distribution with $k=8$ degrees
of freedom is $t$=2.896). 

When the sample is drawn from a bivariate Gaussian distribution with a nonvanishing population
correlation coefficient $\rho$, the Fisher transformation can be used to estimate the confidence
interval for $\rho$ from the measured value $r$. The distribution of $F$,

$F(r) = {1\over 2} \ln \left( {1+r \over 1-r} \right)$

approximately follows a Gaussian distribution with the mean $\mu_F=F(\rho)$ and a standard
deviation $\sigma_F=(N-3)^{-1/2}$. For the above sample with $N=10$ and $r=0.72$, this
approximate approach gives a significance level of 0.8% when $\rho=0$ (instead of the
exact value of 1%).

Pearson's correlation coefficient has two main deficiencies: the measurement
errors for $x$ and $y$ are not used; Pearson's correlation coefficient
is sensitive to outliers 


**Non-parametric coefficients**

Spearman's correlation coefficient and Kendall's correlation coefficient. Based on ranks (the order of a sorted list is called its rank).  The  advantage of ranks is that their distribution is known: each value ($1, \ldots,  N$) occurs exactly once and this fact can be used to derive useful results such as

$\sum_{i=1}^N R_i =  { N \, (N+1) \over 2}$

and

$\sum_{i=1}^N (R_i)^2 =  { N \, (N+1) \, (2N+1)\over 6}.$


Spearman's correlation coefficient is defined similarly to Pearson's coefficient,
but with ranks used instead of data values,

$r_S={\sum_{i=1}^N (R^x_i-\overline{R^x})(R^y_i-\overline{R^y})\over \sqrt{\sum_{i=1}^N (R^x_i-\overline{R^x})^2} \sqrt{\sum_{i=1}^N (R^y_i-\overline{R^y})^2}}$

The distribution of $r_S$ for uncorrelated variables is the same as that for Pearson's 


Other non-parametric approaches include the Kendall's correlation coefficient (see the book), which is based on a comparison of two ranks and does not use differences. Spearman's correlation coefficient is used more often
because it is easier to compute, but Kendall's correlation coefficient is gaining popularity
because it  approaches normality faster than Spearman's correlation coefficient.

These functions are available in SciPy

**General 1D functions**

In the one-dimensional case distribution a simple way to create a sample with a given distribution is to randomly generate a sample over the interval of interest $x_{min}$ to $x_{max}$ and $y_{min}$ to $y_{max}$. For each value of $x$,  if $y$ is smaller than the corresponding $f(x)$ keep the value of $x$ otherwise discard it. A more efficient approach uses the  cumulative distribution function F(x). By fitting a function to F(x) (e.g. a cubic spline) an inverse of the cumulative distribution can be generated. By choosing $y$ from a uniform random sample the value of $x$ can be derived from $F^{-1}(y)$. 

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats, interpolate
from astroML.plotting import hist

#------------------------------------------------------------
# Create a distribution and clone it
Ndata = 1000
Nclone = 100000

#generate function and cumulative data
x = np.linspace(0, 3, Ndata)
y =  np.sin(x)
cumulativey = y.cumsum()
norm = cumulativey[-1]
print norm
cumulativey /= norm

# set up an interpolation of the inverse cumulative distribution
tck = interpolate.splrep(cumulativey, x)

#sample randomly from inverse cumulative distribution
ySample = np.random.random_sample(Nclone)
xSample = interpolate.splev(ySample, tck)


fig = plt.figure(figsize=(20, 10))

#plot function
ax = fig.add_subplot(221)
ax.plot(x,Ndata/3*y/(norm))
ax.set_xlabel('$x$')
ax.set_ylabel('$N(x)$')

#plot cumulative function
ax = fig.add_subplot(222)
ax.plot(x,cumulativey)
ax.set_xlabel('$x$')
ax.set_ylabel('$\Sum N(x)$')


ax = fig.add_subplot(221)
hist(xSample, bins='knuth', ax=ax, histtype='stepfilled', ec='k', fc='#AAAAAA', normed=True)

plt.show()
```

In multidimensional cases, and when the distribution is separable (ie  equal to the product of
independent one-dimensional distributions), one can use the one-dimensional prescription. When the multidimensional distribution is not separable, one needs to consider marginal distributions. For example, in a two-dimensional case $h(x,y)$, one would first draw the value of $x$ using the marginal distribution. Given this $x$ the value of $y$ would be generated using the properly
normalized one-dimensional cumulative conditional probability
distribution in the $y$ direction,

$
        H(y|x) = { \int_{-\infty}^y h(x,y')\,dy' \over  \int_{-\infty}^{\infty} h(x,y')\,dy'}.
$

In higher dimensions, $x$ and $y$ would be kept fixed, and the properly normalized
cumulative distributions of other variables would be used to generate their values.

In the special case of multivariate Gaussian distributions mock samples can
be simply generated in the space of principal axes, and then the values can be ``rotated'' to the appropriate
coordinate system. For example, two independent sets of
values $\eta_1$ and $\eta_2$ can be drawn from an $N(0,1)$ distribution, and then $x$ and $y$ coordinates
can be obtained using the transformations

$
          x = \mu_x + \eta_1 \, \sigma_1 \cos \alpha  - \eta_2 \, \sigma_2 \sin \alpha
$

and

$
          y = \mu_y + \eta_1 \, \sigma_1 \sin \alpha + \eta_2 \, \sigma_2 \cos \alpha.
$
