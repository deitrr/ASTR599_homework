# Part 1: Loading and Exploring Data

Problem 1 - load the data from file.

```python
import numpy as np
data = np.genfromtxt("asteroids5000.csv", delimiter=",")
```

Problem 2 - find the number of missing values and number of complete rows.

```python
nans=np.isnan(data) #find all nans
totalnan = np.sum(nans)
print "Total number of missing values: ",totalnan

nansinrows = nans.sum(1) # number of nans in each row
completerows = np.sum([nansinrows == 0]) # sum of rows with 0 nans
print "Number of complete rows: ",completerows
```

Problem 3 - create new array using only complete rows.

```python
completedata = data[(nansinrows == 0), :] # use mask to select only complete rows
```

Problem 4 - compute max, min, mean, standard deviation of values in each column.

```python
print "Minimum values in each column:\n ",np.min(completedata, 0)
print "Maximum values in each column:\n ",np.max(completedata, 0)
print "Mean values for each column:\n ",np.mean(completedata, 0)
print "Standard deviation for each column:\n ",np.std(completedata, 0)
```

Problem 3 - plot semi-major axis vs. sin(inclination)

```python
import matplotlib.pyplot as plt

# Scatter plot of semi-major axis vs. sin(inclination)
plt.scatter(data[:,0], np.sin(data[:,2]*np.pi/180)) # ANSWER: explanation for this distribution? *******
plt.ylabel("sin(i)")
plt.xlabel("Semi-Major Axis")
```

Problem 4 - color-magnitude diagram

```python
# Color-magnitude diagram
plt.scatter(data[:,3], data[:,5])
plt.plot([2,14], [.78,.78], 'k--') # overplot line that divides two populations (determined by eye)
plt.ylabel("B-V")
plt.xlabel("H")
```

Problem 5 - re-make a vs. sin(i) plot with "families" in color-magnitude diagram color-coded.

```python
bv=data[:,5]
h=data[:,3]

pop1 = data[(bv < 0.78), :] # bluer colors
pop2 = data[(bv >= 0.78), :] # redder colors

plt.scatter(pop1[:,0], np.sin(pop1[:,2]*np.pi/180), c='b',marker='o')
plt.scatter(pop2[:,0], np.sin(pop2[:,2]*np.pi/180), c='r',marker='o')
plt.title('Divided into Populations using CM Diagram')
plt.ylabel("sin(i)")
plt.xlabel("Semi-Major Axis")
```

There is a correlation between color and orbit here; bluer asteroids tend to be found at higher semi-major axes, and asteroids with redder colors tend to be found at smaller semi-major axes. This is consistent with the results of Parker et al. 2008. Their results show both the vertical bands seen here (including the same bands at a = 2.5 and ~2.8-9 as seen here) as well as the correlation between color and semi-major axis.
