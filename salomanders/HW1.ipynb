# Asteroid Data

## Part 1: Loading and Exploring

#### First I read in the data using genfromtxt

```python
import numpy as np
filename = '/astro/users/lmanders/Astro599/2013_fall_ASTR599/notebooks/data/asteroids5000.csv'

f = open(filename)
names = f.readline().strip('#\n').split(',')
print names
f.close()
asteroids = np.genfromtxt(filename, delimiter=',')
```

#### Then I determine the number of missing values

```python
NumberOfMissingValues = np.sum(np.isnan(asteroids))
print NumberOfMissingValues
```

#### Then I determine the number of complete rows

```python
NumColumns       = np.shape(asteroids)[1]
NotNanAsteroid   = np.invert(np.isnan(asteroids))
SumNotNan        = np.sum(NotNanAsteroid, 1)
NumberOfFullRows = np.sum(SumNotNan == NumColumns)
print NumberOfFullRows
```

#### Then I use just those rows to determine the max, min, mean, and std of each column

```python
FullRows     = np.where(SumNotNan == NumColumns)
NewAsteroids = asteroids[FullRows]

MaxValues  = np.max(NewAsteroids, 0)
MinValues  = np.min(NewAsteroids, 0)
MeanValues = np.mean(NewAsteroids, 0)
StdValues  = np.std(NewAsteroids, 0)

padding = 10
print "%s %s %s %s %s" % ('Measure'.ljust(padding),'Max'.ljust(padding), 'Min'.ljust(padding), 'Mean'.ljust(padding), 'Std')
for c1, c2, c3, c4, c5 in zip(names, MaxValues, MinValues, MeanValues, StdValues):
    print "%s %s %s %s %s" % (c1.ljust(padding), c2.astype('|S5').ljust(padding), c3.astype('|S5').ljust(padding), c4.astype('|S5').ljust(padding), c5.astype('|S5'))
```

## Part 2: Plotting

#### As a reminder for what the names of the columns are

```python
OrbitalMeasure = np.genfromtxt(filename, names=names, delimiter=',')
```

#### First I will plot the semi-major axis of each asteroid vs the sine of the inclination angle. Notice the bands of missing asteroids in the plot due to resonances with Jupiter. Asteroids with those semi-major axis are in an unstable orbit and will either be nudged in towards the sun or out towards Jupiter

```python
%pylab inline
import matplotlib.pyplot as plt
plt.scatter(OrbitalMeasure['a'], np.sin(OrbitalMeasure['i']*np.pi/180.), s=5)
plt.xlabel('SemiMajor Axis [AU]')
plt.ylabel('Sin(i)')
```

#### Next I will plot a color-magnitude diagram of the asteroids, as well as a line that divides the population into two families -- red and blue.

```python
%pylab inline
plt.scatter(OrbitalMeasure['BV'], OrbitalMeasure['H'])
plt.plot(np.zeros(20) + 0.775, np.arange(20))
plt.xlabel('B-V')
plt.ylabel('H')
plt.xlim(0.5, 1.1)
plt.ylim(2, 16)
```

#### Just to check that I made the correct split, I plot the two families with their corresponding labeled color

```python
%pylab inline
blue = OrbitalMeasure['BV'] <= 0.775
red  = OrbitalMeasure['BV'] > 0.775
plt.scatter(OrbitalMeasure['BV'][blue], OrbitalMeasure['H'][blue], color='blue',label='Blue')
plt.scatter(OrbitalMeasure['BV'][red], OrbitalMeasure['H'][red], color='red', label='Red')
plt.xlabel('B-V')
plt.ylabel('H')
plt.legend()
```

#### Here I am replotting the first plot of semi-major axis vs sine of the inclination angle, now with colors corresponding to their family in the color-magnitude diagram. Notice that redder asteroids are preferencially found closer to the sun. Their similar colors indicate that they have similar surface chemical compositions. Tying this in with a mild correlation in orbital distance from the sun suggests a common origin.

```python
plt.scatter(OrbitalMeasure['a'][blue], np.sin(OrbitalMeasure['i'][blue]*np.pi/180.), color='blue', s=5)
plt.scatter(OrbitalMeasure['a'][red], np.sin(OrbitalMeasure['i'][red]*np.pi/180.), color='red', s=5)
plt.xlabel('SemiMajor Axis [AU]')
plt.ylabel('Sin(i)')
```
