## PART1

```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
data = np.genfromtxt('asteroids5000.csv', dtype='float', delimiter=',')
x = np.isnan(data)
print 'Total =', data.shape[0] * data.shape[1] #data.size
print 'Number of missing values =', np.isnan(data).sum()
print 'Number of complete rows =', np.sum(np.all(~np.isnan(data),1))
print 'Number of complete columns =', np.sum(np.all(~np.isnan(data),0))

good_rows = np.all(~np.isnan(data),1)
data_new = data[good_rows]
print 'Max =', np.max(data_new, axis=0)
print 'Min =', np.min(data_new, axis=0)
print 'Mean =', np.mean(data_new, axis=0)
print 'Stdv =', np.std(data_new, axis=0)
```
