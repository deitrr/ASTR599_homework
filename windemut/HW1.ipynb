The distribution of asteroids shows clear radial gaps, for instance around 2 AU, 2.5 AU, 2.8 AU, 3.25 AU, and 3.75 AU. These are probably due to orbital resonances; the 2.5 and 2.8 AU clearings correspond to the Kirkwood gaps. Also, there are groups of asteroids which have relatively high inclinations (at 1.9 AU and 2.4 AU). Closer to Jupiter's orbit (~ 5 AU), however, the orbital inclinations are comparatively much lower, perhaps due to the gravitational influence of the massive gas giant.

There are two distinct families of asteroids, one with larger B-V values (redder) and another with smaller B-V values (bluer). The former is mostly found within 3.2 AU while the latter resides throughout the asteroid belt. 

Fig. 3 and 4 of Parker et. al (2008) show that redder asteroids tend to have 2<a/AU<3.2. while bluer asteroids reside beyond a=3.2 AU. This agrees with the trends we see for the color coded sin(i) vs. a plot. The Parker et. al (2008) figures also clearly demonstrate the Kirkwood gaps. The upper envelope of the inclination distribution increases slightly for asteroids in the main belt. This means that asteroids in the outer belt can be more and even more inclined than those in the mid belt and inner belt, respectively.
